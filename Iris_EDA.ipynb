```python
sb.histplot(data=df,x="sepal_length in cm",hue='class',kde=True,bins=10)
plt.title("Sepal_length by class")
plt.show()
```

```python
sb.histplot(data=df,x="sepal_width in cm",hue='class',kde=True,bins=10)
plt.title("Sepal_width by class")
plt.show()
```

```python
sb.histplot(data=df,x="petal_length in cm",hue='class',kde=True,bins=10)
plt.title("Petal_length by class")
plt.show()
```

```python
sb.histplot(data=df,x="petal_width in cm",hue='class',kde=True,bins=10)
plt.title("Petal_width by class")
plt.show()
```
